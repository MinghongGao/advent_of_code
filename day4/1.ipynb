```python
def check_accessible(metrix, x, y):
    width = len(metrix[0])
    height = len(metrix)
    rolls = 0
    for i in range(x - 1, x + 2):
        for j in range(y - 1, y + 2):
            # print(i, j)
            if i < 0 or i >= height or j < 0 or j >= width:
                # print("Out of bounds for", (i, j))
                continue
            if i == x and j == y:
                # print("Same position", (i, j))
                continue

            if metrix[i][j] == "@":
                # print(f"Found roll at ({i},{j})")
                rolls += 1
    # print(f"Position ({x},{y}) has {rolls} rolls")
    if rolls < 4:
        return True
    return False
```

```python

def remove_rolls(metrix):
    total_removed = 0
    for i in range(len(metrix)):
        for j in range(len(metrix[0])):
            if metrix[i][j] == "@":
                if check_accessible(metrix, i, j):
                    metrix[i] = metrix[i][:j] + "." + metrix[i][j + 1 :]
                    total_removed += 1
    return total_removed + remove_rolls(metrix) if total_removed > 0 else 0
```
