```python
def get_invalid_nums_given_range(left: int, right:int) -> list[int]:
    invalid_nums = []
    for num in range(left, right + 1):
        str_num = str(num)
        if len(str_num) % 2 != 0:
            continue

        half_len = len(str_num) // 2
        is_invalid_num = True
        for i in range(half_len):
            if str_num[i] != str_num[i + half_len]:
                is_invalid_num = False

        if is_invalid_num:
            invalid_nums.append(num)
    return invalid_nums
```

```python
def sum_invalid_nums(inp: str) -> int:
    total_sum = 0
    ranges = inp.split(",")
    for r in ranges:
        left, right = r.split("-")
        left, right = int(left), int(right)
        invalid_nums = get_invalid_nums_given_range(left, right)
        total_sum += sum(invalid_nums)
    return total_sum
```
