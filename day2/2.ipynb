```python
def get_invalid_nums_given_range(left: int, right:int) -> list[int]:
    invalid_nums = []
    for num in range(left, right + 1):
        str_num = str(num)

        split_lens = []
        for i in range(len(str_num) - 1, 0,  -1):
            if len(str_num) % i == 0:
                split_lens.append(i)
        
        for split_len in split_lens:
            if "".join(str_num.split(str_num[:split_len])) == "":
                invalid_nums.append(num)
                break
    return invalid_nums
```

```python
def sum_invalid_nums(inp: str) -> int:
    total_sum = 0
    ranges = inp.split(",")
    for r in ranges:
        left, right = r.split("-")
        left, right = int(left), int(right)
        invalid_nums = get_invalid_nums_given_range(left, right)
        total_sum += sum(invalid_nums)
    return total_sum
```
