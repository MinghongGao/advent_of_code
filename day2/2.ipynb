```python
inp = "194-253,81430782-81451118,7709443-7841298,28377-38007,6841236050-6841305978,2222204551-2222236166,2623-4197,318169-385942,9827-16119,580816-616131,646982-683917,147-181,90-120,3545483464-3545590623,4304-5747,246071-314284,8484833630-8484865127,743942-795868,42-53,1435-2086,50480-60875,16232012-16441905,94275676-94433683,61509567-61686956,3872051-4002614,6918792899-6918944930,77312-106847,282-387,829099-1016957,288251787-288311732,6271381-6313272,9877430-10095488,59-87,161112-224439,851833788-851871307,6638265-6688423,434-624,1-20,26-40,6700-9791,990-1307,73673424-73819233"
```

```python
def get_invalid_nums_given_range(left: int, right:int) -> list[int]:
    invalid_nums = []
    for num in range(left, right + 1):
        str_num = str(num)

        split_lens = []
        for i in range(len(str_num) - 1, 0,  -1):
            if len(str_num) % i == 0:
                split_lens.append(i)
        
        for split_len in split_lens:
            if "".join(str_num.split(str_num[:split_len])) == "":
                invalid_nums.append(num)
                break
    return invalid_nums
```

```python
def sum_invalid_nums(inp: str) -> int:
    total_sum = 0
    ranges = inp.split(",")
    for r in ranges:
        left, right = r.split("-")
        left, right = int(left), int(right)
        invalid_nums = get_invalid_nums_given_range(left, right)
        total_sum += sum(invalid_nums)
    return total_sum
```
